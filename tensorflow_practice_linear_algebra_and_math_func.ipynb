tensorflow: math operations and linear algebra 

```python
import tensorflow as tf
```

math lib https://www.tensorflow.org/api_docs/python/tf/math

```python
# absolute values
x = tf.constant([-2.25, 3.25])
tf.abs(x)

# if the numbers are like a + bj , the abs would be sqrt(a^2 + b^2)
```

```python
# addition

x1 = [1, 2, 3, 4, 5]
x2 = [3, 4, 5, 6, 7]

print(tf.add(x1, x2)) #substract, multiply, divide
# tensors can be divided by nan or 0 using: https://www.tensorflow.org/api_docs/python/tf/math/divide_no_nan
```

```python
# elementwise multiplications

x2_single = [9]
print(tf.add(x1, x2_single))
```

```python
# multiplication between different sizes of tensors

x1 = tf.constant([[1, 2, 3, 4, 5, 6]], dtype = tf.float32)
x2 = tf.constant([[1],[2],[3]], dtype = tf.float32)

print(x1.shape)
print(x2.shape)

print(tf.math.multiply(x1, x2))
```

```python
# element wise maximum: Returns the index with the largest value across axes of a tensor.

B = tf.constant([[2, 20, 30, 3, 6], 
                  [3, 11, 16, 1, 8],
                 [14, 45, 23, 5, 27]])
print(B.shape)
print(tf.math.argmax(B,axis=0)) #0: compare column elements (controlling rows)
print(tf.math.argmax(B,axis=1)) #1: compare column elements (controlling columns)
```

```python
# compare tensors to see whether they are equal

x = tf.constant([2, 4])
y = tf.constant(2)
tf.math.equal(x, y)
```

```python
# pow: power of one value to another
x = tf.constant([[2, 2], [3, 3]])
y = tf.constant([[8, 16], [2, 3]])
tf.pow(x, y)  # [[256, 65536], [9, 27]]
```

```python
# reduce_sum: Computes the sum of elements across dimensions of a tensor.

x = tf.constant([[1, 1, 1], [1, 1, 1]]) #shape: 2,3
print(tf.reduce_sum(x).numpy()) # reduce across all dimensions
print(tf.reduce_sum(x, 0).numpy()) # reduce across the first dimension 
# [1, 1, 1] + [1, 1, 1] = [2, 2, 2]
print(tf.reduce_sum(x, 1).numpy()) # reduce across the second dimension -  
# [1, 1] + [1, 1] + [1, 1] = [3, 3]

# also reduce_max, reduce_min, reduce_std, reduce_mean, ...
```

```python
# sigmoid
x = tf.constant([0.0, 1.0, 50.0, 100.0])
print(tf.math.sigmoid(x))
```

```python
# top-k: top k values of an array

result = tf.math.top_k([1, 2, 98, 1, 1, 99, 3, 1, 3, 96, 4, 1],
                        k=3)
print(result.values.numpy())
```

linear algebra
https://www.tensorflow.org/api_docs/python/tf/linalg

```python
# linalg.matmal : matrix multiplication

x1 = tf.constant([[1,2,0],
                 [3,5,-1]])
x2 = tf.constant([[1,2,0],
                 [3,5,-1],
                  [4,5,6]])
print(tf.matmul(x1, x2)) 
print(x1@x2) #reduce way
# notice: (2,3) and (3,2) MUST share the same columns for the first vector ("3"); output size: (2,2)
```

```python
# transpose
tf.transpose(x1)
```

```python
x1_3d = tf.constant([
    [[1,2],[3,5]],

     [[10,2],[1,0]],

      [[5,8],[2,7]]

       ])

x2_3d = tf.constant([
    [[5,1,4],[34,15,1]],

     [[1,2,2],[5,3,2]],

      [[2,2,3],[1,1,1]]

       ])



# how to make batch multiplication, 
# revise the batch of x1_3d to 2,3 ([[1,2],[3,5]]) is called a batch
x1_3d@x2_3d
```

```python
# band_part: Copy a tensor setting everything outside a central band in each innermost matrix to zero.
# cut off the matrix to form different triangles

tensor_two_d = tf.constant([[1,2,0],
                            [3,5,100],
                            [1,5,6],
                            [2,3,8]
                            ], dtype = tf.float16)

#### condition: (lower<0 or m-n <= lower) and (upper>0 or n-m >= upper); 
#### if True, keep value; if False, return 0
# m -- rows, n -- columns

# tf.linalg.band_part(input, 0, -1) ==> Upper triangular part.
# tf.linalg.band_part(input, -1, 0) ==> Lower triangular part.
# tf.linalg.band_part(input, 0, 0) ==> Diagonal.

print(tf.linalg.band_part(tensor_two_d,0,0))

# intermediate computation includes matrices of n-m and m-n
tensor_m_n = tf.constant([
    [0,-1,-2],
    [1,0,-1],
    [2,1,0],
    [3,2,1]
])
tensor_n_m = tf.constant([
    [0,1,2],
    [-1,0,1],
    [-2,-1,0],
    [-3,-2,-1]
])
```

```python
# other topics: cholesky, determinant, cross product, inverse, trace, svd

# inverse: shape must be (n,n)
inv1 = tf.constant([
    [1,-2,0],
    [3,5,100],
    [1,5,6]
], dtype = tf.float16)
tf.linalg.inv(inv1)
```

```python
# svd: a decomposition method to get rid of non-important values
svd_matrix_A = tf.constant([[2, 3], [4, 5], [6, 7]], dtype=tf.float32)
s,u,v = tf.linalg.svd(svd_matrix_A)
print(s)
print(u)
print(v)
```

```python
# einsum operator: Tensor contraction over specified indices and outer product.
# note: always remember the shapes of vectors

import numpy as np

A = np.array([
    [2,6,5,2],
    [2,-2,2,3],
    [1,5,4,0]
])

B = np.array([[2,9,0,3,0],
              [3,6,8,-2,2],
              [1,3,5,0,1],
              [3,0,2,0,5]])

print(A.shape)
print(B.shape)

# compare normal multiplication with eisum
print("matmul C = : \n")
print(A@B)

print("Einsum C = : \n")
print(np.einsum('ij,jk -> ik',A,B))
```

```python
# use eisum to do element wise multiplication

A = np.array([
    [2,6,5,2],
    [2,-2,2,3],
    [1,5,4,0]
])

B = np.array([[2,9,0,3],
              [3,6,8,-2],
              [1,3,5,0]])

print(A*B)
print('\n')

print(np.einsum('ij,ij -> ij',A,B))
```

```python
# use eisum to do transpose

A = np.array([
    [2,6,5,2],
    [2,-2,2,3],
    [1,5,4,0]
])

print(A.T)
print('\n')

print(np.einsum('ij -> ji',A))
```

```python
# einsum: 3d batch size multiplication
# 2,3,4 * 2,4,5 > 2,3,5, b=2

A = np.array([
    [[2,6,5,2],
    [2,-2,2,3],
    [1,5,4,0]],

    [[1,3,1,12],
     [0,2,2,0],
     [1,5,4,1]]
])

B = np.array([[[2,9,0,3,0],
              [3,6,8,-2,2],
              [1,3,5,0,1],
              [3,0,2,0,5]],

              [[1,0,0,3,0],
               [3,0,4,-2,2],
               [1,0,2,0,0],
               [3,0,1,1,0]]
              ])


print(np.matmul(A,B))
print('\n')

print(np.einsum('bij, bjk -> bik',A,B))
```

```python
# einsum: sum all values in a matrix

A = np.array([
    [[2,6,5,2],
    [2,-2,2,3],
    [1,5,4,0]],

    [[1,3,1,12],
     [0,2,2,0],
     [1,5,4,1]]
])

print(np.sum(A))

print(np.einsum('bij ->',A)) # no output in the expression
print(np.einsum('bij -> i',A)) #sum based on row/column
```

```python
# use einsum in tranformer "attention is all you need"
# Q = batchsize, s_q, modelsize
# K = batchsize, s_k, modelsize
Q = np.random.randn(32,64,512)
K = np.random.randn(32,128,512)
np.einsum('bqm,bkm -> bqk',Q,K)
```

```python
# einsum in "reformer: the efficient transformer"
A = np.random.randn(2,4,4,2) #bcij
B = np.random.randn(2,4,4,1) #bcik
print(np.einsum('bcik,bcij -> bckj',B,A).shape) #ik to kj is a transpose
```
